<h3>Data Cleaning</h3>

<h4>1. Find all Null Value in the dataset. If there is any null value in any column, then fill it with the mean of that column.
</h4>

```python
data.isnull().sum()
```

```python
data.dropna(subset=['Make', 'Model','Type','Origin','MSRP','Invoice'],inplace=True)
```

```python
data.isnull().sum()
```

```python
data['Cylinders'].fillna(data['Cylinders'].mean(),inplace=True)
```
